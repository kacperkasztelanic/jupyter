## 2. Automatic Display of the Last Expression

You don't always need `print()`. Jupyter automatically displays the **last expression** in a cell:

```python
# No print() needed — the last expression is displayed automatically
40 + 2
```

```python
# Only the LAST expression is auto-displayed.
# The first expression (10 + 5) is computed but NOT shown.
10 + 5
20 + 3
```

```python
# To display multiple values, use print() for all of them:
print("First:", 10 + 5)
print("Second:", 20 + 3)
```

## 3. Variables Persist Across Cells

Variables defined in one cell are available in all subsequent cells. The notebook maintains a **single shared state** (the kernel).

```python
# Define a variable
message = "I was defined in an earlier cell!"
number = 42
```

```python
# Use variables from the cell above — they persist!
print(message)
print(f"The number is {number}")
```

## 4. Working with Python — Lists, Loops, Functions

You can write any Python code in cells. Here are some common patterns:

```python
# Lists
fruits = ["apple", "banana", "cherry", "date"]
print("Fruits:", fruits)
print("First fruit:", fruits[0])
print("Number of fruits:", len(fruits))
```

```python
# Loops
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
```

```python
# Functions
def greet(name):
    return f"Hello, {name}! Welcome to Jupyter."

greet("World")
```

```python
# List comprehensions
squares = [n ** 2 for n in range(1, 11)]
print("Squares of 1-10:", squares)
```

## 6. Magic Commands

Jupyter has special commands called **magics** that start with `%` (line magic) or `%%` (cell magic). They provide handy shortcuts.

```python
# %timeit — measure execution time of a single statement
%timeit sum(range(1000))
```

```python
# %%timeit — measure execution time of the entire cell
%%timeit
total = 0
for i in range(1000):
    total += i
```

```python
# %who — list all defined variables
%who
```

```python
# %whos — list variables with type and value info
%whos
```

## 8. Rich Output — Beyond Plain Text

Jupyter can display HTML, images, math, and more using IPython's display system.

```python
# Display styled HTML
display(HTML("""
<div style="background-color: #e8f5e9; padding: 15px; border-radius: 8px; border-left: 4px solid #4caf50;">
    <h3 style="color: #2e7d32; margin-top: 0;">✅ Success!</h3>
    <p>This is an HTML box rendered inside a Jupyter notebook.</p>
</div>
"""))
```

```python
# Display Markdown dynamically from code
items = ["Lists", "**Bold text**", "`code`", "*italics*"]
md_text = "### Dynamically Generated Markdown\n\n" + "\n".join(f"- {item}" for item in items)
display(Markdown(md_text))
```

```python
# Display a math equation using LaTeX
display(Math(r"E = mc^2"))
display(Math(r"\sum_{i=1}^{n} i = \frac{n(n+1)}{2}"))
```

## 9. Dictionaries and Data

A common task in notebooks is working with structured data. Here's a taste:

```python
# Working with a dictionary
student = {
    "name": "Alice",
    "age": 22,
    "courses": ["Math", "Physics", "CS"],
    "gpa": 3.8,
}

print(f"Student: {student['name']}")
print(f"Courses: {', '.join(student['courses'])}")
print(f"GPA: {student['gpa']}")
```

```python
# Working with a list of dictionaries (a simple "dataset")
students = [
    {"name": "Alice", "grade": 92},
    {"name": "Bob", "grade": 85},
    {"name": "Charlie", "grade": 78},
    {"name": "Diana", "grade": 95},
    {"name": "Eve", "grade": 88},
]

# Calculate average grade
average = sum(s["grade"] for s in students) / len(students)
print(f"Average grade: {average:.1f}")

# Find the top student
top = max(students, key=lambda s: s["grade"])
print(f"Top student: {top['name']} ({top['grade']})")
```

## 10. Visualizing Data with Built-in Tools

Even without external libraries, you can create simple text-based visualizations.
Later you can install matplotlib, plotly, etc. for real charts.

```python
# Simple text-based bar chart
data = {"Python": 35, "JavaScript": 28, "Rust": 12, "Go": 18, "Java": 22}

max_val = max(data.values())
print("Language Popularity (arbitrary units)\n")
for lang, val in sorted(data.items(), key=lambda x: -x[1]):
    bar = "█" * int(val / max_val * 30)
    print(f"{lang:>12} | {bar} {val}")
```

## 11. Error Handling

When code raises an error, Jupyter shows a **traceback**. This is normal! Read the error message to understand what went wrong.

```python
# This will raise a ZeroDivisionError — that's intentional!
# Read the error message below to see how Jupyter reports errors.
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f"Caught an error: {e}")
    print("Tip: Jupyter shows a full traceback if you don't catch the exception.")
```

## 12. Writing Clean Notebooks

Here are some best practices demonstrated in action:

```python
# ✅ Best Practice: Imports at the top of the notebook
import sys
import os
from datetime import datetime

print("Imports done!")
print(f"Python: {sys.version}")
print(f"Working directory: {os.getcwd()}")
print(f"Current time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```

```python
# ✅ Best Practice: Keep cells focused — one logical operation per cell
# This cell just defines a function
def fibonacci(n):
    """Return the first n Fibonacci numbers."""
    fib = [0, 1]
    for _ in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib[:n]
```

```python
# This cell uses the function
fib_numbers = fibonacci(15)
print(f"First 15 Fibonacci numbers: {fib_numbers}")
```
